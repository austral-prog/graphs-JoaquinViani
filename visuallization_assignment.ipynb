```python
import matplotlib.pyplot as plt
import numpy as np
```

# GRAFICO 1 (LINEA)

```python
x = np.linspace(0, 10, 100)      
y = np.sin(x)     
plt.plot(x, y)
```

Este grafico muestra la funcion seno (senoidal) donde su punto de partida es en (1,0) y sus pico (altos y maximos) es desde 1 y -1 y va desde los valores del 0 al 10

# GRAFICO 2 (BARRAS)

```python
x = np.array(["A","B","C","D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y)
```

Este muestra un grafico de 4 barras de un array generado por numpy y cada barra se le asigna un valor de su frecuencia que viene de otro array generado por numpy.

# GRAFICO 3 (HISTOGRAMA)

```python
data = np.random.gamma(2, 2, 1000) # Gamma distribution
plt.hist(data, bins=30)
```

Es un histogramado generado por duna istribucion gamma  

# GRAFICO 4 (DISPERSION)

```python
x = np.random.normal(0, 1, 100)  # 100 pts, mean=0, std=1
y = x * 2 + np.random.randn(100)
plt.scatter(x, y, c='green', marker='x')
```

Es un grafico de dispersion de datos de una linea que por medio de un ruido, se dispersaron uniformemente y de una manera aleatoria.

# GRAFICO 5 (GRAFICO CIRCULAR)

```python
equipos = "Boca","Riber","Barcelona","Madrid"
frecuencia = [50, 2, 30, 10]

plt.pie(frecuencia, labels = equipos)
```

Este grafico circular respresenta a los equipos mediante labels con sus respectivos frecuencias en cada uno de ellos, formando asi un circulo con la informacion brindada
